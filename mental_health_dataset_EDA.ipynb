## ***SECTION - 2.1: DROPPING NAN COLUMNS TO REMOVE REDUNDENCY***


## Purpose
Identify and remove columns containing **'nan' or '_Nan'** in their names from the dataset. This step is crucial because, especially in the context of one-hot encoded categorical features, a "_nan" column often signifies redundancy. For instance, if a feature like `mh_coverage` is one-hot encoded into `mh_coverage_0.0` and `mh_coverage_1.0`, and both are 0, it implicitly means `mh_coverage_nan` is 1. Therefore, explicitly retaining the _nan column would introduce perfect (or near-perfect) inverse correlation and redundancy with the other categories. Removing these columns avoids multicollinearity, simplifies the dataset, and prepares the data for more robust analysis by eliminating implicitly represented or highly correlated information.

## Conclusion
A total of 11 columns, including `mh_family_hist_nan` and `mh_coverage_flag_nan`, that contained 'nan' or '_nan' in their names were successfully identified and dropped. This action was taken to remove redundant information, particularly from one-hot encoded features where the absence of other categories already implies the 'nan' state. By removing these highly correlated and implicitly represented _nan columns, the dataset's dimensions were reduced from (1235, 117) to (1235, 106), resulting in a cleaner `df_eda_view_nonan` DataFrame. This ensures that subsequent analyses are based on more independent features, improving model stability and interpretability for HR insights.

```python
# Identify and drop columns with 'nan' or '_NaN' in their names
import pandas as pd
import numpy as np

# If df_eda_view does not exist, the next line will raise a NameError
df_cleaned_nan_cols = df_eda_view.copy()

nan_like_cols = [col for col in df_cleaned_nan_cols.columns if 'nan' in col.lower() or '_nan' in col.lower()]

if nan_like_cols:
    print(f"Identified {len(nan_like_cols)} columns with 'nan' or '_NaN' in their names:")
    print(nan_like_cols)
    df_cleaned_nan_cols = df_cleaned_nan_cols.drop(columns=nan_like_cols)
    print(f"\nDropped these columns. New shape: {df_cleaned_nan_cols.shape}")
else:
    print("No columns with 'nan' or '_NaN' in their names found.")

# Display the first few rows of the cleaned dataframe
print("\nFirst 5 rows of the dataframe after dropping 'nan' columns:")
display(df_cleaned_nan_cols.head())

df_eda_view_nonan = df_cleaned_nan_cols.copy()
df_eda_view_nonan_2 = df_cleaned_nan_cols.copy()
print("\nCleaned dataframe stored in df_eda_view_nonan.")
```

### ***SECTION 2.2.2 FEATURE SELECTION***

### Purpose
Apply VarianceThreshold for feature selection to remove features with low variance, which are near-constant and provide little information for modeling. This step aims to reduce dimensionality and improve model performance by eliminating uninformative features.

### Conclusion
VarianceThreshold with a threshold of 0.01 was applied to the `df_eda_view_nonan` DataFrame. This process identified and removed 2 near-constant features (e.g., `'mh_not_eff_treat_impact_on_work_1.0'`, `'sex_Prefer not to say'`), resulting in `df_eda_view_feature_selection` with 104 features. The remaining features, representing more variability, were retained for further analysis.

```python
# ===== Feature selection using VarianceThreshold =====
# Output: df_eda_view_feature_selection (DataFrame) and selected_features list

import numpy as np
import pandas as pd
from sklearn.feature_selection import VarianceThreshold

# 0) Source check
try:
    df_src = df_eda_view_nonan.copy()
except NameError:
    raise RuntimeError("df_eda_view_nonan not found. Create df_eda_view_nonan before running this cell.")

print("Source shape:", df_src.shape)

# 1) Define feature candidate columns (using same list used for PCA earlier)
# re-use explicit column lists from above but filter to present columns to avoid KeyErrors.
col_num = ['age']

col_binary = [
 'mh_bad_response_workplace_-1.0','mh_bad_response_workplace_0.0','mh_bad_response_workplace_1.0',
 'mh_conseq_coworkers_-1','mh_conseq_coworkers_0.0','mh_conseq_coworkers_1.0',
 'mh_coverage_flag_-1.0','mh_coverage_flag_0.0','mh_coverage_flag_1.0',
 'mh_diagnos_proffesional',
 'mh_employer_discussion_-1.0','mh_employer_discussion_0.0','mh_employer_discussion_1.0',
 'mh_family_hist_-1.0','mh_family_hist_0.0','mh_family_hist_1.0',
 'mh_resources_provided_-1.0','mh_resources_provided_0.0','mh_resources_provided_1.0',
 'mh_sought_proffes_treatm','prev_employers_flag','self_empl_flag',
 'tech_comp_flag_-1','tech_comp_flag_0.0','tech_comp_flag_1.0','tech_flag'
]

col_likert = [
 'future_mh_specification','future_ph_specification','mh_discussion_cowork','mh_discussion_neg_impact',
 'mh_discussion_supervis','mh_disorder_current','mh_disorder_past','mh_eff_treat_impact_on_work',
 'mh_medical_leave_0.0','mh_medical_leave_1.0','mh_medical_leave_2.0','mh_medical_leave_3.0','mh_medical_leave_4.0','mh_medical_leave_5.0',
 'mh_not_eff_treat_impact_on_work_0.0','mh_not_eff_treat_impact_on_work_1.0','mh_not_eff_treat_impact_on_work_2.0',
 'mh_not_eff_treat_impact_on_work_3.0','mh_not_eff_treat_impact_on_work_4.0',
 'mh_sharing_friends/fam_flag_1.0','mh_sharing_friends/fam_flag_2.0','mh_sharing_friends/fam_flag_3.0',
 'mh_sharing_friends/fam_flag_4.0','mh_sharing_friends/fam_flag_5.0',
 'ph_discussion_neg_impact','remote_flag'
]

col_ordinal = [
 'comp_no_empl','prev_mh_benefits_0.0','prev_mh_benefits_1.0','prev_mh_benefits_2.0','prev_mh_benefits_3.0',
 'prev_mh_conseq_coworkers','prev_mh_discuss_neg_conseq_0.0','prev_mh_discuss_neg_conseq_1.0','prev_mh_discuss_neg_conseq_2.0','prev_mh_discuss_neg_conseq_3.0',
 'prev_mh_discussion','prev_mh_discussion_cowork','prev_mh_discussion_supervisor','prev_mh_importance_employer_0.0','prev_mh_importance_employer_1.0',
 'prev_mh_importance_employer_2.0','prev_mh_importance_employer_3.0','prev_mh_resources','prev_ph_discuss_neg_conseq'
]

col_nominal = [
 'country_live_Australia','country_live_Canada','country_live_Germany','country_live_Netherlands','country_live_UK','country_live_USA',
 'country_work_Australia','country_work_Canada','country_work_Germany','country_work_Netherlands','country_work_UK','country_work_USA',
 'mh_hurt_on_career_Maybe','mh_hurt_on_career_No, I don\'t think it would','mh_hurt_on_career_No, it has not',
 'mh_hurt_on_career_Yes, I think it would','mh_hurt_on_career_Yes, it has',
 'mh_neg_view_cowork_Maybe','mh_neg_view_cowork_No, I don\'t think they would','mh_neg_view_cowork_No, they do not',
 'mh_neg_view_cowork_Yes, I think they would','mh_neg_view_cowork_Yes, they do',
 'role_designer','role_dev','role_developer','role_devops','role_lead','role_leadership','role_support','role_sysadmin',
 'sex_Female','sex_Male','sex_Other','sex_Prefer not to say'
]

# 2) Filter only present columns
candidate_cols = col_num + col_binary + col_likert + col_ordinal + col_nominal
present_cols = [c for c in candidate_cols if c in df_src.columns]
missing_cols = [c for c in candidate_cols if c not in df_src.columns]
print(f"\nCandidate count: {len(candidate_cols)}, Present in df: {len(present_cols)}")
if missing_cols:
    print("Warning: these expected columns are missing and will be skipped (example):", missing_cols[:12])

# 3) Build feature matrix for variance threshold
X = df_src[present_cols].copy()

# Ensure numeric dtype (VarianceThreshold requires numeric array)
# If any columns are object due to bad encoding, try to coerce numeric (safe attempt)
for c in X.columns:
    if X[c].dtype == 'O':
        X[c] = pd.to_numeric(X[c], errors='coerce')

# If any remaining NaNs were introduced by coercion, fill them with 0 (OHE columns should be 0/1)
X = X.fillna(0)

print("\nPrepared matrix shape for VarianceThreshold:", X.shape)

# 4) VarianceThreshold: remove near-constant features.
# Threshold selection: 0.01 -> features with variance <= 0.01 are removed. Adjust if needed.
threshold = 0.01
vt = VarianceThreshold(threshold=threshold)
vt.fit(X)
mask = vt.get_support()        # boolean mask of features to keep
selected_features = list(X.columns[mask])
removed_features = list(X.columns[~mask])

print(f"\nVarianceThreshold (th={threshold}) kept {len(selected_features)} features; removed {len(removed_features)} features.")
if removed_features:
    print("removed features:", removed_features[:20])

# 5) Create output DataFrame (selected features only)
df_eda_view_feature_selection = df_src[selected_features].copy()
print("\ndf_eda_view_feature_selection shape:", df_eda_view_feature_selection.shape)

# 6) Quick summaries of selected features
print("\nTop 12 features by variance (descending):")
variances = X[selected_features].var().sort_values(ascending=False)
display(variances.head(12))

# 7) Save the lists to variables for downstream use
selected_features_list = selected_features
removed_features_list = removed_features


print(" - df_eda_view_feature_selection (DataFrame with selected features)")
print(" - selected_features_list (list)")
print(" - removed_features_list (list)")
```

## ***SECTION - 3.1 : PCA AND CLUSTERING (KMEANS, GMM)***

### ***SECTION - 3.1.1: PCA(STANDARD-SCALING)ON`df_eda_view_feature_scaling`***

### Purpose
Apply Principal Component Analysis (PCA) after standard scaling to reduce the dimensionality of the `df_eda_view_feature_selection` data. This step aims to transform the original, potentially correlated, features into a new set of orthogonal (uncorrelated) variables called principal components. The goal is to capture the maximum variance in the data with a fewer number of components, providing a lower-dimensional representation that is easier to visualize and interpret for subsequent analysis.

### Conclusion
After applying `StandardScaler` to ensure all features contribute equally regardless of their original scale, PCA was performed on the 104 feature-selected attributes. The analysis indicated that 61 principal components are required to explain approximately 95.22% of the total variance in the dataset. For visualization purposes, the first two principal components (PC1 and PC2) were extracted, collectively explaining about 19.72% of the variance. A 2D scatter plot along with correlation heatmap  was generated to visualize the data in this reduced space, and the top 20 features contributing most significantly to PC1, PC2 and PC3 (based on their loadings) were identified, offering an initial glimpse into the underlying structures represented by these components.

### HR Insights
Standard scaling is critical here to ensure that all HR-related attributes, regardless of their measurement scale (e.g., age in years vs. binary flags), contribute fairly to the identification of underlying patterns. PCA allows HR to move beyond individual survey questions and discover *latent dimensions* or overarching themes within employee sentiment and experience. For example, PC1 might represent an 'Overall Workplace Stress' factor, driven by features like perceived negative impact of mental health discussion, absence due to mental health, and general discomfort. PC2 could represent 'Perceived Company Support', weighted by availability of resources, positive employer discussion, and previous positive experiences with benefits. These latent dimensions offer a powerful, simplified lens for understanding complex employee data without relying on specific clusters at this stage.

### HR Solutions
By identifying these abstract principal components, HR can effectively prioritize strategic interventions. If 'Overall Workplace Stress' (PC1) is a dominant factor, HR can explore broad initiatives like stress management programs, workload balancing, or promoting work-life integration. If 'Perceived Company Support' (PC2) is weak, HR can focus on increasing awareness of existing mental health benefits, enhancing communication channels, and training managers to be more supportive. This dimensionality reduction helps HR leaders communicate key issues and solutions to senior management more concisely, moving from managing dozens of metrics to focusing on a few impactful drivers of employee well-being and engagement, even before specific employee segments are identified.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn for heatmap
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


try:
    df_pca_standard = df_eda_view_feature_selection.copy()
except NameError:
    raise RuntimeError("df_eda_view_feature_selection not found. Please run the cell that creates it.")

# --- 1. Scale the data using StandardScaler ---
print("Scaling data with StandardScaler...")
scaler = StandardScaler()
data_scaled_standard = scaler.fit_transform(df_pca_standard)
print("Scaling complete. Scaled data shape:", data_scaled_standard.shape)

# --- 2. Perform PCA ---
# PCA for 95% variance
pca_95_standard = PCA(n_components=0.95, random_state=42)
pca_95_standard.fit(data_scaled_standard)
n_components_95_standard = pca_95_standard.n_components_
explained_variance_95_standard = np.sum(pca_95_standard.explained_variance_ratio_)
print(f"\nPCA (StandardScaler): {n_components_95_standard} components needed for {explained_variance_95_standard:.4f} (95% target) explained variance.")

# PCA for 3 components (for loadings heatmap and 2D visualization)
pca_3_standard = PCA(n_components=3, random_state=42)
pca_3_standard_coords = pca_3_standard.fit_transform(data_scaled_standard)
explained_variance_3_standard = np.sum(pca_3_standard.explained_variance_ratio_)

print(f"PCA (StandardScaler): First 3 components explain {explained_variance_3_standard:.4f} of the variance.")
print(f"PCA (StandardScaler): First 2 components explain {np.sum(pca_3_standard.explained_variance_ratio_[:2]):.4f} of the variance.")


# --- 3. Visualize the 2D PCA projection (using the first two components from pca_3_standard_coords) ---
plt.figure(figsize=(9, 7))
plt.scatter(pca_3_standard_coords[:, 0], pca_3_standard_coords[:, 1], alpha=0.55, s=30)
plt.title("2D PCA Projection (StandardScaler) - Before Clustering")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()

# --- 4. Generate PCA Loadings Heatmap for top features across PC1-PC3 ---
print("\nGenerating PCA Loadings Heatmap for PC1, PC2, and PC3...")
try:
    feature_names_standard = df_pca_standard.columns
    # Create a DataFrame for loadings from the 3 components
    loadings_standard = pd.DataFrame(pca_3_standard.components_.T, index=feature_names_standard, columns=['PC1', 'PC2', 'PC3'])

    # Calculate absolute max loading across PC1, PC2, PC3 for each feature
    loadings_standard['max_abs'] = loadings_standard[['PC1', 'PC2', 'PC3']].abs().max(axis=1)

    # Select top N features by their maximum absolute loading
    top_n_features = 20 # You can adjust this number
    top_loadings = loadings_standard.sort_values('max_abs', ascending=False).head(top_n_features).drop(columns='max_abs')

    plt.figure(figsize=(10, 8))
    sns.heatmap(top_loadings, annot=True, cmap='RdBu_r', center=0, fmt=".3f", linewidths=.5, cbar_kws={"shrink": 0.75})
    plt.title(f"PCA Loadings Heatmap (Top {top_n_features} features for PC1, PC2, PC3)")
    plt.xlabel("Principal Component")
    plt.ylabel("Feature")
    plt.tight_layout()
    plt.show()

    print(f"\nTop {top_n_features} features contributing to PC1, PC2, and PC3:")
    display(top_loadings)

except Exception as e:
    print(f"\nCould not display loadings heatmap: {e}")

print("\nStandardScaler PCA analysis complete.")
```

### ***SECTION - 3.1.2: PCA(HYBRID SCALE), KMEANS AND GMM ON `df_eda_view_feature_selection`***

### Purpose

This section aims to reduce the dimensionality of the feature-selected data (`df_eda_view_feature_selection`) using Principal Component Analysis (PCA) and then identify natural groupings within this reduced space using KMeans and Gaussian Mixture Models (GMM). A hybrid scaling approach is employed to appropriately handle different data types before PCA, ensuring a more robust and meaningful dimensionality reduction. The goal is to uncover distinct employee segments based on their mental health-related attributes, which can then be used for targeted HR interventions.

---

### HR Insights

1.  **Justification for Hybrid Scaling:** While `StandardScaler` (z-score normalization) is common, a hybrid approach was necessary for this mixed-type dataset. `StandardScaler` was used for the purely numerical `age` column. `MinMaxScaler` (0-1 scaling) was applied to Likert-scale features, preserving their ordinal relationships while making them comparable to binary features. `RobustScaler` was used for other ordinal features (like `comp_no_empl`), as it handles outliers better by using medians and interquartile ranges, which is crucial for skewed survey data. Binary and one-hot encoded nominal features were passed through as-is, as their 0/1 representation is already suitable. This tailored scaling ensures that different types of HR-related data contribute appropriately to PCA without disproportionate influence.

2.  **PCA Explained Variance & Loadings Interpretation:**
    * **Interpretation:** The first two principal components (PC1 and PC2), which were used for clustering, together explain **26.43%** of the variance (PC1: 14.06%, PC2: 12.37%). The first three components explain 33.05%. While this may seem low, it is common in high-dimensional survey datasets where many factors contribute independently.
    * **Loadings (from Heatmap):** The PCA loadings heatmap reveals the drivers for these components:
        * **PC1 (14.06% variance):** This component is dominated by high positive loadings from `mh_disorder_current` (0.428), `mh_disorder_past` (0.425), `mh_sought_proffes_treatm` (0.407), and `mh_diagnos_proffesional` (0.354). This component clearly captures a dimension related to the **"History and Current Experience of Mental Health Conditions."** A high score on PC1 indicates a greater history and current experience with diagnosed mental health issues.
        * **PC2 (12.37% variance):** This component is driven by perceptions of the workplace. It has high *positive* loadings for `mh_bad_response_workplace_0.0` (0.366) and `mh_conseq_coworkers_0.0` (0.361), which represent "No, I did not experience a bad response." It has high *negative* loadings for all the "-1.0" flags, such as `mh_coverage_flag_-1.0` (-0.316) and `mh_resources_provided_-1.0` (-0.297), which represent "I don't know." This component thus represents **"Awareness and Perception of Workplace Support."** A high score means "Aware of resources and perceives a positive/supportive culture," while a low score means "Unaware of company resources" (which strongly correlates with being self-employed).

3.  **KMeans (k=4) Justification:**
    * **Elbow and Silhouette Scores:** For KMeans, the Silhouette plot showed a distinct, high peak at **k=4 (0.7311)**, indicating the best separation and cluster cohesion at this number. This was significantly higher than the scores for k=5 (0.4949) and k=6 (0.4434).
    * **Comparison to GMM:** GMM's best Silhouette score was at **k=3 (0.7185)**. While also a strong score, the KMeans(k=4) solution provides a more granular and highly interpretable segmentation. As the profiling tables show, k=4 cleanly separates the population into four *perfectly distinct* personas based on `self_empl_flag` and `prev_employers_flag`. The GMM(k=3) solution, by contrast, merged two of these distinct groups. Therefore, **KMeans with k=4 was chosen** as the superior model for generating actionable HR insights.

---

### KMeans Cluster Interpretation (k=4)

The clustering revealed four extremely distinct groups, primarily defined by employment status (`self_empl_flag`) and prior work experience (`prev_employers_flag`).

* **Cluster 1 (N=895, "The Core Established Workforce")**
    * **Size:** The largest cluster (72.5% of respondents).

```python
# ===== Fixed: PCA + KMeans + GMM on df_eda_view_feature_selection (handles missing cols) =====
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings("ignore")

# 0) using df_eda_view_feature_selection
try:
    df_orig = df_eda_view_feature_selection
except NameError:
    raise RuntimeError("df_eda_view_feature_selection not found. Create df_eda_view_feature_selection before running this cell.")
df = df_orig.copy()

# 1) expected column groups - same names as before
col_num = ['age']

col_binary = [
 'mh_bad_response_workplace_-1.0','mh_bad_response_workplace_0.0','mh_bad_response_workplace_1.0',
 'mh_conseq_coworkers_-1','mh_conseq_coworkers_0.0','mh_conseq_coworkers_1.0',
 'mh_coverage_flag_-1.0','mh_coverage_flag_0.0','mh_coverage_flag_1.0',
 'mh_diagnos_proffesional',
 'mh_employer_discussion_-1.0','mh_employer_discussion_0.0','mh_employer_discussion_1.0',
 'mh_family_hist_-1.0','mh_family_hist_0.0','mh_family_hist_1.0',
 'mh_resources_provided_-1.0','mh_resources_provided_0.0','mh_resources_provided_1.0',
 'mh_sought_proffes_treatm','prev_employers_flag','self_empl_flag',
 'tech_comp_flag_-1','tech_comp_flag_0.0','tech_comp_flag_1.0','tech_flag'
]

col_likert = [
 'future_mh_specification','future_ph_specification','mh_discussion_cowork','mh_discussion_neg_impact',
 'mh_discussion_supervis','mh_disorder_current','mh_disorder_past','mh_eff_treat_impact_on_work',
 'mh_medical_leave_0.0','mh_medical_leave_1.0','mh_medical_leave_2.0','mh_medical_leave_3.0','mh_medical_leave_4.0','mh_medical_leave_5.0',
 'mh_not_eff_treat_impact_on_work_0.0','mh_not_eff_treat_impact_on_work_2.0',
 'mh_not_eff_treat_impact_on_work_3.0','mh_not_eff_treat_impact_on_work_4.0',
 'mh_sharing_friends/fam_flag_1.0','mh_sharing_friends/fam_flag_2.0','mh_sharing_friends/fam_flag_3.0',
 'mh_sharing_friends/fam_flag_4.0','mh_sharing_friends/fam_flag_5.0',
 'ph_discussion_neg_impact','remote_flag'
]

col_ordinal = [
 'comp_no_empl','prev_mh_benefits_0.0','prev_mh_benefits_1.0','prev_mh_benefits_2.0','prev_mh_benefits_3.0',
 'prev_mh_conseq_coworkers','prev_mh_discuss_neg_conseq_0.0','prev_mh_discuss_neg_conseq_1.0','prev_mh_discuss_neg_conseq_2.0','prev_mh_discuss_neg_conseq_3.0',
 'prev_mh_discussion','prev_mh_discussion_cowork','prev_mh_discussion_supervisor','prev_mh_importance_employer_0.0','prev_mh_importance_employer_1.0',
 'prev_mh_importance_employer_2.0','prev_mh_importance_employer_3.0','prev_mh_resources','prev_ph_discuss_neg_conseq'
]

col_nominal = [
 'country_live_Australia','country_live_Canada','country_live_Germany','country_live_Netherlands','country_live_UK','country_live_USA',
 'country_work_Australia','country_work_Canada','country_work_Germany','country_work_Netherlands','country_work_UK','country_work_USA',
 'mh_hurt_on_career_Maybe','mh_hurt_on_career_No, I don\'t think it would','mh_hurt_on_career_No, it has not',
 'mh_hurt_on_career_Yes, I think it would','mh_hurt_on_career_Yes, it has',
 'mh_neg_view_cowork_Maybe','mh_neg_view_cowork_No, I don\'t think they would','mh_neg_view_cowork_No, they do not',
 'mh_neg_view_cowork_Yes, I think they would','mh_neg_view_cowork_Yes, they do',
 'role_designer','role_dev','role_developer','role_devops','role_lead','role_leadership','role_support','role_sysadmin',
 'sex_Female','sex_Male','sex_Other'
]

# 2) keep only columns that exist in df (prevents KeyError)
all_expected = col_num + col_binary + col_likert + col_ordinal + col_nominal
present_cols = [c for c in all_expected if c in df.columns]
missing_cols = [c for c in all_expected if c not in df.columns]

print(f"Expected columns: {len(all_expected)}")
print(f"Columns present in df_eda_view_feature_selection: {len(present_cols)}")
if missing_cols:
    print(f"\nWARNING: {len(missing_cols)} expected columns are MISSING (they will be skipped).")
    print("Example missing (up to 20):", missing_cols[:20])

# rebuild per-group lists using only present columns
col_num = [c for c in col_num if c in present_cols]
col_binary = [c for c in col_binary if c in present_cols]
col_likert = [c for c in col_likert if c in present_cols]
col_ordinal = [c for c in col_ordinal if c in present_cols]
col_nominal = [c for c in col_nominal if c in present_cols]

feature_cols = col_num + col_binary + col_likert + col_ordinal + col_nominal
print(f"\nUsing {len(feature_cols)} columns for PCA after filtering missing columns.")

# 3) ColumnTransformer with present cols only
ct = ColumnTransformer(transformers=[
    ('num', StandardScaler(), col_num) if col_num else ('num', 'passthrough', []),
    ('binary', 'passthrough', col_binary),
    ('likert', MinMaxScaler(), col_likert),
    ('ordinal', RobustScaler(), col_ordinal),
    ('nominal', 'passthrough', col_nominal)
], remainder='drop', sparse_threshold=0)

X_scaled = ct.fit_transform(df[feature_cols])
print("Scaled matrix shape:", X_scaled.shape)

# 4) PCA on scaled features (3 components)
pca = PCA(n_components=3, random_state=42)
X_pca = pca.fit_transform(X_scaled)
df_pca_coords = pd.DataFrame(X_pca, columns=['PCA1','PCA2','PCA3'], index=df.index)
explained_123 = np.sum(pca.explained_variance_ratio_)
print(f"PCA explained variance PC1 = {pca.explained_variance_ratio_[0]:.4f}")
print(f"PCA explained variance PC2 = {pca.explained_variance_ratio_[1]:.4f}")
print(f"PCA explained variance PC3 = {pca.explained_variance_ratio_[2]:.4f}")
print(f"PCA explained variance PC1+PC2+PC3 = {explained_123:.4f}")
print(f"Number of components used to capture 95% variance: {n_components_95_standard}")

# ADDITION: Plot 2D PCA projection alone, before clustering
plt.figure(figsize=(9, 7))
plt.scatter(df_pca_coords['PCA1'], df_pca_coords['PCA2'], alpha=0.55, s=30)
plt.title("2D PCA Projection (Hybrid Scale) Before Clustering")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()

# ---- ADDITION: PCA loadings heatmap (top features) ----
# compute loadings and plot heatmap of top-N features by max absolute loading across PC1-PC3
try:
    loadings = pd.DataFrame(pca.components_.T, index=feature_cols, columns=['PC1','PC2','PC3'])
    loadings['max_abs'] = loadings[['PC1','PC2','PC3']].abs().max(axis=1)
    top_n = 20
    top_feats = loadings.sort_values('max_abs', ascending=False).head(top_n).drop(columns='max_abs')
    plt.figure(figsize=(10,8))
    sns.heatmap(top_feats, annot=True, cmap='RdBu_r', center=0, fmt=".3f")
    plt.title(f"PCA loadings (top {top_n} features by |loading| across PC1-PC3)")
    plt.tight_layout()
    plt.show()
except Exception as e:
    print("Could not create PCA heatmap:", e)
# ---- end heatmap block ----

# 5) Elbow & Silhouette for KMeans (on PCA1,PCA2)
Ks = list(range(2,11))
inertias = []
silhouettes = []
for k in Ks:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
    labels = kmeans.fit_predict(df_pca_coords[['PCA1','PCA2']])
    inertias.append(kmeans.inertia_)
    sil = silhouette_score(df_pca_coords[['PCA1','PCA2']], labels)
    silhouettes.append(sil)
    print(f"KMeans k={k} -> inertia={kmeans.inertia_:.2f}, silhouette={sil:.4f}")

# plots
fig, axs = plt.subplots(1,2, figsize=(14,4))
axs[0].plot(Ks, inertias, '-o'); axs[0].set_title('Elbow (KMeans inertia)'); axs[0].set_xlabel('k'); axs[0].grid(True)
axs[1].plot(Ks, silhouettes, '-o'); axs[1].set_title('Silhouette (KMeans)'); axs[1].set_xlabel('k'); axs[1].grid(True)
plt.show()

best_k_kmeans = Ks[int(np.nanargmax(silhouettes))]
print("Best k (KMeans) by silhouette:", best_k_kmeans)

# 6) Fit final KMeans and store labels
kmeans_final = KMeans(n_clusters=best_k_kmeans, random_state=42, n_init=20)
labels_km = kmeans_final.fit_predict(df_pca_coords[['PCA1','PCA2']])
df_clusters = df.copy()
df_clusters['Cluster_KMeans'] = labels_km

# plot KMeans clusters on PCA2
plt.figure(figsize=(8,6))
scatter = plt.scatter(df_pca_coords['PCA1'], df_pca_coords['PCA2'], c=labels_km, cmap='tab10', s=35, alpha=0.8)
plt.title(f'KMeans (k={best_k_kmeans}) on PCA(2)')
plt.xlabel('PCA1'); plt.ylabel('PCA2'); plt.grid(True)
plt.legend(*scatter.legend_elements(), title="clusters")
plt.show()

# 7) GMM: Elbow (proxy) & Silhouette (on PCA2)
sil_gmm = []
inertia_proxy = []
for k in Ks:
    gmm = GaussianMixture(n_components=k, covariance_type='full', random_state=42)
    labs = gmm.fit_predict(df_pca_coords[['PCA1','PCA2']])
    sil = silhouette_score(df_pca_coords[['PCA1','PCA2']], labs)
    sil_gmm.append(sil)
    # a proxy for "inertia" for GMM: negative log-likelihood (lower is better) -> we store -score
    inertia_proxy.append(-gmm.score(df_pca_coords[['PCA1','PCA2']]))

    print(f"GMM k={k} -> silhouette={sil:.4f}, neg-log-lik proxy={inertia_proxy[-1]:.2f}")

fig, axs = plt.subplots(1,2, figsize=(14,4))
axs[0].plot(Ks, inertia_proxy, '-o'); axs[0].set_title('Elbow proxy (GMM neg-log-lik)'); axs[0].set_xlabel('k'); axs[0].grid(True)
axs[1].plot(Ks, sil_gmm, '-o'); axs[1].set_title('Silhouette (GMM)'); axs[1].set_xlabel('k'); axs[1].grid(True)
plt.show()

best_k_gmm = Ks[int(np.nanargmax(sil_gmm))]
print("Best k (GMM) by silhouette:", best_k_gmm)

# 8) Fit final GMM and store labels
gmm_final = GaussianMixture(n_components=best_k_gmm, covariance_type='full', random_state=42)
labels_gmm_final = gmm_final.fit_predict(df_pca_coords[['PCA1','PCA2']])
df_clusters['Cluster_GMM'] = labels_gmm_final

plt.figure(figsize=(8,6))
scatter = plt.scatter(df_pca_coords['PCA1'], df_pca_coords['PCA2'], c=labels_gmm_final, cmap='tab10', s=35, alpha=0.8)
plt.title(f'GMM (k={best_k_gmm}) on PCA(2)')
plt.xlabel('PCA1'); plt.ylabel('PCA2'); plt.grid(True)
plt.legend(*scatter.legend_elements(), title="clusters")
plt.show()

# 9) Cluster interpretation — include self_empl_flag and produce both numeric means and binary proportions
print("\n--- Cluster sizes (KMeans) ---")
print(df_clusters['Cluster_KMeans'].value_counts().sort_index())
print("\n--- Cluster sizes (GMM) ---")
print(df_clusters['Cluster_GMM'].value_counts().sort_index())

# Prepare lists for profiling
numeric_profile_cols = [c for c in (['age','comp_no_empl','mh_discussion_neg_impact','ph_discussion_neg_impact']) if c in df.columns]
binary_profile_cols = [c for c in col_binary + col_nominal if c in df.columns]

def profile_clusters(df_in, cluster_col):
    print(f"\n--- Profiling by {cluster_col} ---")
    # numeric means
    if numeric_profile_cols:
        print("\nNumeric means per cluster:")
        display(df_in.groupby(cluster_col)[numeric_profile_cols].mean().round(3))
    # binary/nominal proportions (mean of 0/1 columns)
    if binary_profile_cols:
        print("\nBinary/nominal proportions per cluster (top 12 by overall prevalence):")
        proportions = df_in.groupby(cluster_col)[binary_profile_cols].mean().T
        # order by overall prevalence to show important flags first
        top = proportions.mean(axis=1).sort_values(ascending=False).index[:12]
        display(proportions.loc[top].round(3))
    # show self_empl_flag specifically (counts)
    if 'self_empl_flag' in df_in.columns:
        print("\nself_empl_flag counts per cluster:")
        display(df_in.groupby(cluster_col)['self_empl_flag'].value_counts(normalize=False).unstack(fill_value=0))

# profile KMeans
profile_clusters(df_clusters, 'Cluster_KMeans')
# profile GMM
profile_clusters(df_clusters, 'Cluster_GMM')

# store outputs for later use
# df_pca_coords contains PCA coordinates
# df_clusters contains original features + Cluster_KMeans + Cluster_GMM
print("\nDone. Results kept in df_pca_coords and df_clusters.")
```

```python
# =====Feature Engineering (Cleaned: removed Resilience_Index, Age_Adjusted_Resilience, Remote_Binary)=====

import numpy as np
import pandas as pd

# ---------- A. Source check ----------
if 'df_eda_view_feature_selection' not in globals():
    raise RuntimeError("df_eda_view_feature_selection not found. Create it (feature-selection DF) and run this cell again.")

df_src = df_eda_view_feature_selection.copy()
print("Source (feature-selection) shape:", df_src.shape)

# ---------- B. Helper ----------
def has(col):
    return col in df_src.columns

# ---------- C. Create engineered features ----------
df_eng = df_src.copy()
created_cols_list = []
original_cols_used_for_eng = []

# 1) Coverage_Index
cov_candidates = []
if has('mh_coverage_flag_1.0'):
    cov_candidates.append('mh_coverage_flag_1.0')
if has('mh_resources_provided_1.0'):
    cov_candidates.append('mh_resources_provided_1.0')

if cov_candidates:
    df_eng['Coverage_Index'] = df_eng[cov_candidates].fillna(0).astype(float).sum(axis=1)
    created_cols_list.append('Coverage_Index')
    original_cols_used_for_eng.extend(cov_candidates)
    print("Created Coverage_Index from:", cov_candidates)
else:
    print("Coverage_Index NOT created.")

# 2) Openness_Index
openness = [c for c in ['mh_discussion_cowork', 'mh_discussion_supervis'] if has(c)]
if openness:
    df_eng['Openness_Index'] = df_eng[openness].astype(float).mean(axis=1)
    created_cols_list.append('Openness_Index')
    original_cols_used_for_eng.extend(openness)
    print("Created Openness_Index from:", openness)
else:
    print("Openness_Index NOT created.")

# 3) Company_Support (depends on Coverage_Index only)
if has('prev_mh_resources') and 'Coverage_Index' in df_eng.columns:
    max_val = df_eng['prev_mh_resources'].max()
    if pd.isna(max_val) or max_val == 0:
        prev_norm = df_eng['prev_mh_resources'].fillna(0).astype(float)
    else:
        prev_norm = df_eng['prev_mh_resources'].fillna(0).astype(float) / float(max_val)
    df_eng['Company_Support'] = prev_norm * df_eng['Coverage_Index']
    created_cols_list.append('Company_Support')
    original_cols_used_for_eng.append('prev_mh_resources')
    print("Created Company_Support.")
else:
    print("Company_Support NOT created.")

# Remove duplicate originals
original_cols_used_for_eng = list(set(original_cols_used_for_eng))

# ---------- D. Drop original columns used for engineering ----------
cols_to_drop = [col for col in original_cols_used_for_eng if col in df_eng.columns]
if cols_to_drop:
    print(f"\nDropping {len(cols_to_drop)} original columns:", cols_to_drop)
    df_eng = df_eng.drop(columns=cols_to_drop)
else:
    print("No original-engineering columns to drop.")

# ---------- E. Categorize engineered features ----------
engineered_numeric = []
engineered_binary = []

for col in created_cols_list:
    if col in df_eng.columns:
        if set(df_eng[col].dropna().unique()).issubset({0, 1}):
            engineered_binary.append(col)
        else:
            engineered_numeric.append(col)

print("\nCategorized Engineered Columns:")
print("  Numeric/Continuous:", engineered_numeric)
print("  Binary:", engineered_binary)

# ---------- F. Report ----------
new_cols = created_cols_list
print("\nEngineered columns created:", new_cols)
print("Final df shape:", df_eng.shape)

if new_cols:
    print("\nDescriptive stats:")
    display(df_eng[new_cols].describe().T.round(3))

# ---------- G. Cluster-aware summaries ----------
possible_cluster_cols = [c for c in df_eng.columns if 'cluster' in c.lower()]
if possible_cluster_cols:
    cluster_col = possible_cluster_cols[0]
    print("\nUsing cluster column:", cluster_col)
    print(df_eng[cluster_col].value_counts().sort_index())
    print("\nEngineered-feature means per cluster:")
    display(df_eng.groupby(cluster_col)[new_cols].mean().round(3))
else:
    print("\nNo cluster column found inside df_eda_view_feature_eng.")

# ---------- H. Output ----------
df_eda_view_feature_eng = df_eng.copy()
print("\nDone — df_eda_view_feature_eng created.")
```

## ***SECTION-4.3: ANALYSIS AND VISUALIZATION OF ENGINEERED FEATURES WITH `df_eda_view_feature_eng_2`***


### 1. Definition of Engineered Features
To simplify the analysis, we synthesized related survey questions into four high-level aggregated features. Here is how they are calculated:

* **`Coverage_Index`**
    * *Definition:* Measures the aggregate awareness of mental health benefits.
    * *Formula:* Sum of positive indicators ('Yes') for `mh_coverage_flag` and `mh_resources_provided`.
    * *Logic:* `Coverage_Index = Sum(Coverage_Flag + Resources_Provided)`

* **`Openness_Index`**
    * *Definition:* Measures the psychological safety and comfort level of the workplace culture.
    * *Formula:* The average score of comfort levels when discussing mental health with coworkers and supervisors.
    * *Logic:* `Openness_Index = Mean(Discussion_Coworkers, Discussion_Supervisors)`

* **`Company_Support`**
    * *Definition:* A weighted score representing the depth of institutional support, combining current benefit awareness with past resource history. it tells "Am I an experienced beneficiary of corporate support systems?".
    
    * *Formula:* Normalized score of `prev_mh_resources` multiplied by the `Coverage_Index`.
    * *Logic:* `Company_Support = Normalized_Prev_Resources * Coverage_Index`
---

### 2. Purpose
The objective of this section is to map the four employee personas identified by **KMeans (k=4)** against these new engineered metrics. By moving from individual survey questions to these aggregate indices, we can draw clearer insights into the behavioral patterns and support gaps that define each cluster.

---

### 3. Cluster Interpretation via Engineered Features

Applying these metrics to our established clusters reveals distinct behavioral profiles:

* **Cluster 1 (The Core Established Workforce):**
    * **Behavior:** This group scores the highest on **`Company_Support`** and **`Openness_Index`**.
    * **Insight:** They represent the ideal state—employees who are not only aware of their benefits (`Coverage_Index`) but feel supported by the institution and comfortable discussing mental health.

* **Cluster 0 (The Young, New Workforce):**
    * **Behavior:** While this group has a decent **`Coverage_Index`** (they know benefits exist), their **`Company_Support`** score is drastically lower (often near zero).
    * **Insight:** This highlights a "Usage Gap." They are aware of the safety net but, likely due to a lack of tenure or onboarding, have not effectively engaged with or felt supported by these resources.

* **Clusters 2 & 3 (The Self-Employed):**
    * **Behavior:** As expected, these groups score **0.0** on `Company_Support` and `Coverage_Index` because they lack an employer. However, they consistently show the highest values **`remote_flag` columns**.
    * **Insight:** These clusters trade institutional stability for high location flexibility. Their high values of remote_flag validate their identification as the "Remote/Contractor" segments.

### 4. Conclusion
The engineered features successfully validate our segmentation. They quantify the **institutional gap** between established employees and new hires (Cluster 1 vs. Cluster 0) and clearly show the **structural difference** of the self-employed workforce (Clusters 2 & 3) through the remote_flag column.

```python
# Create respondent-level table: engineered columns + cluster assignment
# -------------------------------------------------------------------
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# --- 0) CONFIG: change k if different KMeans cluster count is needed ---
K_KMEANS = 4
ENGINEERED_COLS = [
    'Coverage_Index', 'Openness_Index',
    'Company_Support'
]

# --- 1) Prepare engineered dataframe (df_eda_view_feature_eng_2) ---
# If already present, use it. Otherwise try to build it from df_eda_view_feature_selection.
try:
    df_eng2 = df_eda_view_feature_eng_2.copy()
    print("Using existing df_eda_view_feature_eng_2 (shape: {})".format(df_eng2.shape))
except NameError:
    # Fallback if df_eda_view_feature_eng_2 is not found
    try:
        _src = df_eda_view_feature_selection.copy()
    except NameError:
        raise RuntimeError("Neither 'df_eda_view_feature_eng_2' nor 'df_eda_view_feature_selection' found. Create them first.")
    print("df_eda_view_feature_eng_2 not found — creating from df_eda_view_feature_selection.")

    # Create engineered columns safely
    def safe_has(c): return c in _src.columns

    df_eng2 = _src.copy()
    created = []

    # Coverage_Index
    cov = [c for c in ['mh_coverage_flag_1.0', 'mh_resources_provided_1.0'] if safe_has(c)]
    if cov:
        df_eng2['Coverage_Index'] = df_eng2[cov].fillna(0).astype(float).sum(axis=1)
        created.append('Coverage_Index')

    # Openness_Index
    open_cols = [c for c in ['mh_discussion_cowork', 'mh_discussion_supervis'] if safe_has(c)]
    if open_cols:
        df_eng2['Openness_Index'] = df_eng2[open_cols].astype(float).mean(axis=1)
        created.append('Openness_Index')

    # Company_Support
    if safe_has('prev_mh_resources') and 'Coverage_Index' in df_eng2.columns:
        maxv = df_eng2['prev_mh_resources'].replace([np.inf, -np.inf], np.nan).fillna(0).max()
        if maxv == 0:
            prev_norm = df_eng2['prev_mh_resources'].fillna(0).astype(float)
        else:
            prev_norm = df_eng2['prev_mh_resources'].fillna(0).astype(float) / float(maxv)
        df_eng2['Company_Support'] = prev_norm * df_eng2['Coverage_Index']
        created.append('Company_Support')


    # final check: ensure all engineered cols exist (if not, they will be absent)
    present_engineered = [c for c in ENGINEERED_COLS if c in df_eng2.columns]
    print("Engineered columns created/present:", present_engineered)
    # assign back to global name so subsequent cells can use it
    df_eda_view_feature_eng_2 = df_eng2.copy()
    print("Created df_eda_view_feature_eng_2 (shape {}).".format(df_eda_view_feature_eng_2.shape))

# --- 2) Determine cluster labels for each respondent ---
cluster_col_name = 'Cluster_KMeans'
cluster_labels = None # Initialize to None
_df_clusters_ref = None # Reference to df_clusters if it exists

if cluster_col_name in df_eng2.columns:
    print("Cluster column already present in df_eng2; using it.")
    cluster_labels = df_eng2[cluster_col_name].copy()
else: # Only proceed if cluster_labels haven't been set yet from df_eng2
    # Attempt to load df_clusters from global scope if it exists
    try:
        _df_clusters_ref = df_clusters # If df_clusters is defined, assign its reference
    except NameError:
        # df_clusters is not defined yet, _df_clusters_ref remains None
        pass

    if _df_clusters_ref is not None and cluster_col_name in _df_clusters_ref.columns:
        print("Mapping cluster labels from existing df_clusters.")
        cluster_labels = _df_clusters_ref[cluster_col_name].reindex(df_eng2.index)
    else:
        # compute clusters freshly (safe fallback)
        try:
            df_fs = df_eda_view_feature_selection.copy()
        except NameError:
            raise RuntimeError("df_eda_view_feature_selection not found; cannot compute clusters.")

        print("Computing PCA(2) + KMeans(k={}) on df_eda_view_feature_selection to get cluster labels.".format(K_KMEANS))

        # build a numeric matrix from df_fs: coerce non-numeric -> numeric where possible, fillna=0
        X = df_fs.copy()
        for c in X.columns:
            if X[c].dtype == 'O':
                X[c] = pd.to_numeric(X[c], errors='coerce')
        X = X.fillna(0).values

        # scale + PCA(2)
        scaler = StandardScaler()
        Xs = scaler.fit_transform(X)
        pca = PCA(n_components=2, random_state=42)
        Xp = pca.fit_transform(Xs)

        # fit KMeans
        km = KMeans(n_clusters=K_KMEANS, random_state=42, n_init=20)
        labels_km = km.fit_predict(Xp)

        # create a small df to map labels back to original index
        df_pca_indexed = pd.DataFrame(Xp, index=df_fs.index, columns=['PCA1','PCA2'])
        df_pca_indexed['Cluster_KMeans'] = labels_km
        # reindex or map into df_eng2
        cluster_labels = df_pca_indexed['Cluster_KMeans'].reindex(df_eng2.index)

# Ensure cluster_labels is defined at this point
if cluster_labels is None:
    raise RuntimeError("Failed to determine cluster labels after all attempts.")

# Define df_clusters for downstream use (always create it here using the determined labels)
try:
    df_clusters = df_eda_view_feature_selection.copy()
    df_clusters['Cluster_KMeans'] = cluster_labels
except NameError:
    raise RuntimeError("df_eda_view_feature_selection not found; cannot create df_clusters for downstream use.")

# final cluster labels (ensure no NaNs; if NaNs, fill with -1)
cluster_labels = cluster_labels.reindex(df_eng2.index)
if cluster_labels.isnull().any():
    print("Warning: some respondents have missing cluster labels; assigning -1 for missing.")
    cluster_labels = cluster_labels.fillna(-1).astype(int)
else:
    cluster_labels = cluster_labels.astype(int)

# attach cluster labels to engineered dataframe (create a respondent-level view)
cols_present_engineered = [c for c in ENGINEERED_COLS if c in df_eng2.columns]
if not cols_present_engineered:
    raise RuntimeError("No engineered columns found in df_eda_view_feature_eng_2. Nothing to attach.")

df_engineered_with_clusters = df_eng2[cols_present_engineered].copy()
df_engineered_with_clusters.insert(0, 'Cluster_KMeans', cluster_labels.values)
# keep original index; optionally reset index for a respondent id column
df_engineered_with_clusters = df_engineered_with_clusters.reset_index().rename(columns={'index':'respondent_index'})

# --- 3) Summary statistics by cluster (counts + means) ---
cluster_counts = df_engineered_with_clusters['Cluster_KMeans'].value_counts().sort_index()
cluster_means = df_engineered_with_clusters.groupby('Cluster_KMeans')[cols_present_engineered].mean().round(3)

print("\nRespondents per cluster (KMeans):")
print(cluster_counts.to_string())

print("\nEngineered-feature means per cluster (KMeans):")
display(cluster_means)

# --- 4) Also create a compact profile dataframe for easy export / reporting ---
cluster_profile = cluster_means.copy()
cluster_profile['n_members'] = cluster_counts
cluster_profile = cluster_profile.reset_index().rename(columns={'Cluster_KMeans':'cluster'})

# Save to globals for use in notebook
df_engineered_with_clusters = df_engineered_with_clusters.copy()
cluster_profile = cluster_profile.copy()

print("Output variables available:")
print(" - df_engineered_with_clusters : respondent-level table with engineered cols + cluster (shape {})".format(df_engineered_with_clusters.shape))
print(" - cluster_profile : cluster-level summary with counts and means (shape {})".format(cluster_profile.shape))

# Example: peek first 6 rows
display(df_engineered_with_clusters.head(6))
```

## ***SECTION - 5.6: DISTRIBUTION OF MH TREATMENT SEEKING RATE BY CLUSTER***

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure df_engineered_with_clusters is available and has Cluster_KMeans
try:
    df_plot_clustered = df_engineered_with_clusters.copy()
except NameError:
    raise RuntimeError("df_engineered_with_clusters not found. Please run previous cells to create it.")

if 'Cluster_KMeans' not in df_plot_clustered.columns:
    raise RuntimeError("Cluster_KMeans column not found in df_engineered_with_clusters. Cannot plot by cluster.")

# Ensure df_eda_view_feature_selection is available for the original feature
try:
    df_original_flags = df_eda_view_feature_selection.copy()
except NameError:
    raise RuntimeError("df_eda_view_feature_selection not found. Please run previous cells to create it.")

# Merge the original 'mh_sought_proffes_treatm' column into the clustered dataframe
# Assuming 'respondent_index' in df_plot_clustered corresponds to the index of df_original_flags
if 'respondent_index' in df_plot_clustered.columns:
    df_plot_clustered = df_plot_clustered.set_index('respondent_index')

if 'mh_sought_proffes_treatm' not in df_original_flags.columns:
    raise RuntimeError("'mh_sought_proffes_treatm' column not found in df_eda_view_feature_selection. Cannot plot.")

df_plot_clustered = df_plot_clustered.merge(df_original_flags[['mh_sought_proffes_treatm']], left_index=True, right_index=True, how='left')

# Calculate the average 'mh_sought_proffes_treatm' for each Cluster_KMeans
# Since it's a binary (0/1) flag, the mean represents the proportion/rate
average_treatment_seeking_rate = df_plot_clustered.groupby('Cluster_KMeans')['mh_sought_proffes_treatm'].mean().reset_index()
average_treatment_seeking_rate.rename(columns={'mh_sought_proffes_treatm': 'Treatment_Seeking_Rate'}, inplace=True)

print("--- Average Mental Health Treatment Seeking Rate by KMeans Cluster ---")
display(average_treatment_seeking_rate.round(3))

# Visualize the average rates
plt.figure(figsize=(9, 6))
sns.barplot(x='Cluster_KMeans', y='Treatment_Seeking_Rate', data=average_treatment_seeking_rate, palette='rocket')
plt.title('Average Mental Health Treatment Seeking Rate by KMeans Cluster')
plt.xlabel('KMeans Cluster')
plt.ylabel('Treatment Seeking Rate (Proportion)')
plt.ylim(0, 1) # Rates are proportions, so they should be between 0 and 1

# Add percentages on top of the bars
for index, row in average_treatment_seeking_rate.iterrows():
    plt.text(row.name, row.Treatment_Seeking_Rate + 0.02, f'{row.Treatment_Seeking_Rate:.1%}', color='black', ha="center")

plt.tight_layout()
plt.show()
```
